### Ejercicio 2

En este ejercicio nos gustaría ver si podemos agrupar los alimentos en grupos diferentes según sus características de una manera `no supervisada`. Con esa finalidad, utilizaremos el método `KMeans` que hemos visto en la teoría. **(3 puntos)** 

a) Antes de aplicar el KMeans, normaliza los datos utilizando la función `MinMaxScaler`. ¿ Por que razón es importante hacer esta transformación? <span style="font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px; ">NM</span> **(1 punto)**



```python
from sklearn.preprocessing import MinMaxScaler

# Respuesta
X = np.nan_to_num(compositionDf.loc[:, valuesToPot].values)

scaler = MinMaxScaler()
XNEW = scaler.fit_transform(X)

# De esta manera podemos agrupar i transformar los datos en valores sin tanta diferencia, ya que seran valores de entre 0 i 1
```

b) Uno de los problemas de los métodos no supervisados es identificar el número óptimo de clusters. Para poder estimar esta número óptimo, se utiliza frecuentemente `el método de Elbow`. Busca información sobre este método y utilízalo. ¿ Cual es el número óptimo de clusters? <span style="font-family: Courier New; background-color: #f2ae72; color: #000000; padding: 3px; ">EI</span> **(1 punto)** 

```python
from sklearn.cluster import KMeans

# Respuesta
distortions = []
distortionsNew = []
K = range(1,15)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(X)
    distortions.append(kmeanModel.inertia_)

    kmeanModel.fit(XNEW)
    distortionsNew.append(kmeanModel.inertia_)

plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method without MinMaxScaler')
plt.show()

plt.plot(K, distortionsNew, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method with MinMaxScaler')
plt.show()

# Como podemos ver al usar el MinMaxScaler tenemos un resultado mas exacto.
# Mirando el segundo grafico podemos cojer como valor k valores entre el 10 i el 12 asi que escojere el 11 como numero de clusters
```

c) Aplica el método KMeans especificando el número óptimo de clusters obtenido en el apartado anterior. Compara las agrupaciones obtenidas con la categoría de la columna `Category`. Podemos encontrar alguna relación entre el clustering del KMeans y les categorias (Cereals, Vegetables, Meat, etc) de la base? <span style="font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px; ">NM</span> **(1 punto)** 

```python
# Respuesta
kmeanModel = KMeans(n_clusters=11)
label = kmeanModel.fit_predict(X)

u_labels = np.unique(label)

plt.figure(figsize=(10, 10))
for i in u_labels:
    plt.scatter(X[label == i , 0] , X[label == i , 1] , label = i)
plt.legend()
plt.show()

c_labels = np.unique(compositionDf['Category'])
plt.figure(figsize=(10, 10))
for i in c_labels:
    plt.scatter(X[compositionDf['Category'] == i , 0] , X[compositionDf['Category'] == i , 1] , label = i)
plt.legend()
plt.show()

# Lo que si podemos decir es que el numero de clusters es correcto y respecto a la relacion que hay entre 
# las dos podemos ver que no en su totalidad pero si que ha agupado muchos de ellos correctamente
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics

# Respuesta
x = titanicDf[['Pclass', 'Age', 'Fare']]
y = titanicDf['Survived']
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, shuffle=True)

clf = RandomForestClassifier(n_estimators=1000)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

# Model accuracy
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```

(f) **(Opcional)** En los apartados anteriores hemos visto la importancia de separar nuestro dataset en el subconjunto de entrenamiento y evaluación. Sin embargo, no podemos asegurar que nuestro modelo sea muy generalizable ya que sólo lo hemos probado por una partición de train/test concreta. Para mejorar la validación de nuestro modelo, podemos utilizar lo que se conoce como **validación cruzada**.
(1) Busca información sobre el método k-Fold y explica en qué consiste. (2) A continuación aplica un k-Fold con k=10. ¿Podemos concluir que nuestro modelo es bueno para predecir la supervivencia de los pasajeros?  <span style="font-family: Courier New; background-color: #ffcc5c; color: #000000; padding: 2px; ">EG</span>

### References

K-Fold Cross-Validation in Python Using SKLearn

```python
# Respuesta

# (1)
# Este metodo se utiliza para ver los datos los cuales entrenando el modelo pueda dar el
# mejor resultado y se utiliza k para repartir los datos en k grupos con los quales hace
# pruebas entre ellos.

# (2)
# Como veremos en el siguiente código vemos que hay resultados de una exactitud
# razonable que van del 70% hasta poco más del 80%, pero otros están también por el 50%,
# como conclusión podríamos decir que es un modelo que funciona bastante bien con
# este dataset, pero no tiene la suficiente fiabilidad dependiendo el 'fold', así que
# si utilizamos como entrenamiento el modelo con mayor exactitud podríamos decir que
# sí que es un modelo razonable que podamos utilizar.
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score

x = np.array(titanicDf[['Pclass', 'Age', 'Fare']])
y = np.array(titanicDf['Survived'])

clf = RandomForestClassifier(n_estimators=1000)

acc_score = []

kf = KFold(n_splits=10)
for train_index, test_index in kf.split(x):
    X_train, X_test = x[train_index], x[test_index]
    y_train, y_test = y[train_index], y[test_index]

    clf.fit(X_train,y_train)
    pred_values = clf.predict(X_test)
     
    acc = accuracy_score(pred_values , y_test)
    acc_score.append(acc)

print('Accuracy of each fold:', acc_score)
```
